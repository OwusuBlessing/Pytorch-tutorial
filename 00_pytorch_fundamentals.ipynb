### Reshaping ,Stacking, Squeezing and unsqueezing
 - reshape --change shape
 - view - returna view of input tensor but keep the same memor as the original tensor
 -stacking - combining multiple tensors on top of each other
 - Squeezing - removind all "1" dimension
 - unsqueez - add a "1" dimension to target tensor
 - permute - return a view the input with dimensions permuted swapped in a certain way

```python
#create a tensor
x = torch.arange(1,10)
```

```python
x.shape
```

```python
#add extra dimesnion
x_reshape1 = x.reshape(1,9)
x_reshape1
```

```python
x_reshape2 = x.reshape(3,3)
x_reshape2
```

### Reproducibility

```python

t1 = torch.rand(3,3)
t2 = torch.rand(3,3)
t1 , t2
```

```python
print(t1 == t2)
```

```python
from torch.random import manual_seed
#make random and reproducibles tensor
#set seed
random_seed = 42
torch.manual_seed(random_seed)
tc = torch.rand(3,4)
torch.manual_seed(random_seed)
td = torch.rand(3,4)
print(tc == td)
```
